```python
#@title One‑Tap Telegram Login (Auto‑Submit)
#@markdown 1. Enter your phone number (with country code) in the box below.
#@markdown 2. A login code will be sent to your Telegram app.
#@markdown 3. Enter the code in the box that appears.
#@markdown 4. If you have 2FA, you'll be asked for your password.
#@markdown 5. Once logged in, the session string is **automatically sent** to your bot.
#@markdown    Then tap the **Open Bot** button to return to the chat.

!pip install -q pyrogram
!pip install -q pyrogram tgcrypto

from pyrogram import Client
from IPython.display import display, HTML

API_ID = 24274245
API_HASH = "3b07c776229e28264bf0d71056a49cb1"
BOT_USERNAME = "forward_Restricted23bot"

phone = input("📱 Phone number (with + CountryCode): ").strip()

app = Client(":memory:", api_id=API_ID, api_hash=API_HASH, in_memory=True)

async def main():
    await app.connect()
    sent = await app.send_code(phone)
    code = input("📩 Login Telegram code: ").strip()
    try:
        await app.sign_in(phone, sent.phone_code_hash, code)
    except Exception:
        password = input("🔒 2FA password: ").strip()
        await app.check_password(password)

    session_str = await app.export_session_string()

    # Automatically send the session string to your bot
    await app.send_message(BOT_USERNAME, session_str)

    # Show a button to open the bot
    bot_link = f"https://t.me/{BOT_USERNAME}"
    display(HTML(f'''
        <h3>✅ Session sent to the bot!</h3>
        <p>Tap the button below to return to Telegram. Your bot is ready to use.</p>
        <a href="{bot_link}" style="font-size:18px;padding:10px;background:#0088cc;color:white;text-decoration:none;border-radius:5px;">
            📱 Open Bot
        </a>
    '''))

    await app.disconnect()

await main()
```
